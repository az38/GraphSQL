```python
from pyspark.sql.types import *

server = '<server>.database.windows.net'
database = '<database>'
username = '<username>'
password = '<password>'
jdbc_url = f'jdbc:sqlserver://{server};databaseName={database};user={username};password={password};'

squad_df = spark.read \
    .format("jdbc") \
    .option("url", jdbc_url) \
    .option("query", """
        select 
            players.id as id, 
            players.name as name, 
            clubs.id as club, 
            squad.year as year,
            squad.joined as joined
        from
            Squad, Players, Clubs 
        where MATCH(Clubs-(Squad)->Players)
    """) \
    .option("customSchema", "id INT, name STRING, club INT, year INT, joined INT") \
    .load() 

squad_df.write \
    .format('com.databricks.spark.csv') \
    .mode("overwrite") \
    .option("mergeSchema", "true") \
    .saveAsTable("squad_csv")

squad_df.write \
    .format('delta') \
    .mode("overwrite") \
    .option("mergeSchema", "true") \
    .saveAsTable("squad_delta")

players_df = spark.read \
    .format("jdbc") \
    .option("url", jdbc_url) \
    .option("dbtable", "players") \
    .load() 

players_df.select("ID", "Name") \
    .write \
    .format('com.databricks.spark.csv') \
    .mode("overwrite") \
    .saveAsTable("players_csv")

players_df.select("ID", "Name") \
    .write \
    .format('delta') \
    .mode("overwrite") \
    .saveAsTable("players_delta")
"""
clubs_df = spark.read \
    .format("jdbc") \
    .option("url", jdbc_url) \
    .option("dbtable", "clubs") \
    .load() 

clubs_df.select("ID", "Name") \
    .write \
    .format('com.databricks.spark.csv') \
    .mode("overwrite") \
    .saveAsTable("clubs_csv")

clubs_df.select("ID", "Name") \
    .write \
    .format('delta') \
    .mode("overwrite") \
    .saveAsTable("clubs_delta")
"""
```
